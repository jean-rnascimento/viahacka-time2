```python
#carga do dataset lojista

df_lojista = spark.read\
  .format('org.apache.spark.sql.execution.datasources.csv.CSVFileFormat')\
  .option('header', 'true')\
  .load(cos.url('data_asset/lojista_csv_afxfkwe161qozdy9feukl9yqp.csv', 'projetoviagrupo2-donotdelete-pr-y2sctck2rd1okn'))
```

```python
#regra de negocio para excluir compras nao aprovadas

df_compra_apv = df_compra.filter(col('dataaprovacao').isNotNull())
```

```python
#join compra e compraentrega

condicao = [df_compra_apv.idcompra == df_compraentrega.idcompra,
            df_compra_apv.idbandeira == df_compraentrega.idbandeira]

df_compra_join_compraentrega = df_compra_apv.join(df_compraentrega, condicao, 'inner').select(df_compra_apv.idcompra,
                                                                                          df_compra_apv.idbandeira,
                                                                                          df_compra_apv.dataaprovacao,
                                                                                          df_compraentrega.idcompraentrega,
                                                                                          df_compraentrega.idlojista)
```

```python
#join da resultante de compra e compraentrega com a tabela de lojista

condicao = [df_compra_join_compraentrega.idlojista == df_lojista.idlojista,
            df_compra_join_compraentrega.idbandeira == df_lojista.idbandeira]

df_compra_join_compraentrega_join_lojista = df_compra_join_compraentrega.join(df_lojista, condicao, 'inner').select(df_compra_join_compraentrega.idcompra,
                                                                                                                    df_compra_join_compraentrega.idbandeira,
                                                                                                                    df_compra_join_compraentrega.idcompraentrega,
                                                                                                                    df_compra_join_compraentrega.idlojista,
                                                                                                                    df_compra_join_compraentrega.dataaprovacao,
                                                                                                                    df_lojista.estado)
```

```python
#join da resultante de compra, compraentrega e lojista com a tabela de compraentregasku e eliminar compras com valor negativo

condicao = [df_compra_join_compraentrega_join_lojista.idcompraentrega == df_compraentregasku.idcompraentrega,
            df_compra_join_compraentrega_join_lojista.idbandeira      == df_compraentregasku.idbandeira]

df_sug_preco = df_compra_join_compraentrega_join_lojista.join(df_compraentregasku, condicao, 'inner').select(df_compra_join_compraentrega_join_lojista.idcompra,
                                                                                                             df_compra_join_compraentrega_join_lojista.idbandeira,
                                                                                                             df_compra_join_compraentrega_join_lojista.idcompraentrega,
                                                                                                             df_compra_join_compraentrega_join_lojista.idlojista,
                                                                                                             df_compra_join_compraentrega_join_lojista.estado,
                                                                                                             df_compra_join_compraentrega_join_lojista.dataaprovacao,
                                                                                                             df_compraentregasku.idsku.alias('compraentregasku_idsku'),
                                                                                                             df_compraentregasku.valorvendaunidade)

df_sug_preco = df_sug_preco.filter(col('valorvendaunidade') > 0)
```

```python
#regra de negocio para pegar apenas a ultima compra de cada sku x logista

df_sug_preco_ult_cmp = df_sug_preco.groupBy('idbandeira', 'idlojista', 'estado', 'compraentregasku_idsku').agg(max('dataaprovacao').alias('dataaprovacao'))
```

```python
#no resultado da ultima compra de cada sku x lojista coloca devolta no dataset os dados de valor e idcompra

condicao = [df_sug_preco.idbandeira             == df_sug_preco_ult_cmp.idbandeira,
            df_sug_preco.idlojista              == df_sug_preco_ult_cmp.idlojista,
            df_sug_preco.estado                 == df_sug_preco_ult_cmp.estado,
            df_sug_preco.compraentregasku_idsku == df_sug_preco_ult_cmp.compraentregasku_idsku,
            df_sug_preco.dataaprovacao          == df_sug_preco_ult_cmp.dataaprovacao]

df_sug_preco = df_sug_preco.join(df_sug_preco_ult_cmp, condicao, 'inner').select(df_sug_preco_ult_cmp.idbandeira,
                                                                                 df_sug_preco_ult_cmp.idlojista,
                                                                                 df_sug_preco_ult_cmp.estado,
                                                                                 df_sug_preco_ult_cmp.compraentregasku_idsku.alias('df_sug_preco_idsku'),
                                                                                 df_sug_preco_ult_cmp.dataaprovacao,
                                                                                 df_sug_preco.idcompra,
                                                                                 df_sug_preco.valorvendaunidade)

df_sug_preco = df_sug_preco.groupBy('idbandeira', 'idlojista', 'estado', 'df_sug_preco_idsku', 'dataaprovacao', 'idcompra', 'valorvendaunidade').count()
```

```python
#com a listagem das ultima compras de todos os sku x lojista pega o maior e o menor preco de cada sku

df_sku_preco_max_min = df_sug_preco.select('df_sug_preco_idsku', 'valorvendaunidade').groupBy('df_sug_preco_idsku').agg(max('valorvendaunidade').alias('max_valorvendaunidade'), min('valorvendaunidade').alias('min_valorvendaunidade'))
```

```python
#adiciona ao dataset principal os valores de min e max de cada sku

condicao = [df_sug_preco.df_sug_preco_idsku == df_sku_preco_max_min.df_sug_preco_idsku]

df_resultante = df_sug_preco.join(df_sku_preco_max_min, condicao, 'inner').select(df_sug_preco.idbandeira,
                                                                                  df_sug_preco.idlojista,
                                                                                  df_sug_preco.estado,
                                                                                  df_sug_preco.df_sug_preco_idsku,
                                                                                  df_sug_preco.valorvendaunidade,
                                                                                  df_sku_preco_max_min.max_valorvendaunidade,
                                                                                  df_sku_preco_max_min.min_valorvendaunidade)
```

```python
#enrriquecimento final do dataset com campos de valor medio do preco, percentual de ocilacao do preco e se esta praticando o maior ou menor preco da plataforma

df_resultante = df_resultante.select(col('idbandeira').cast('int').alias('IDBANDEIRA'),
                                     col('idlojista').cast('int').alias('IDLOJISTA'),
                                     col('estado').cast('string').alias('ESTADO'),
                                     col('df_sug_preco_idsku').cast('int').alias('IDSKU'),
                                     col('valorvendaunidade').cast('float').alias('VALORVENDAUNIDADE'),
                                     col('max_valorvendaunidade').cast('float') .alias('MAX_VALORVENDAUNIDADE'),
                                     col('min_valorvendaunidade').cast('float') .alias('MIN_VALORVENDAUNIDADE'),
                                     ((col('max_valorvendaunidade')+
                                       col('min_valorvendaunidade'))/2).cast('float').alias('MEDIA_VALORVENDAUNIDADE'),
                                     (round(((col('valorvendaunidade')/
                                              ((col('max_valorvendaunidade')+
                                                col('min_valorvendaunidade'))/2))-1)*100,2)).cast('float').alias('PERC_OCILACAO'),
                                     when(col('valorvendaunidade') == col('min_valorvendaunidade'), 'S')
                                      .otherwise('N').cast('string').alias('MENOR_PRECO'),
                                     when(col('valorvendaunidade') == col('max_valorvendaunidade'), 'S')
                                      .otherwise('N').cast('string').alias('MAIOR_PRECO'))
```

```python
#grava os dados no DB2

df_resultante.write.option("createTableColumnTypes", "ESTADO VARCHAR(2), MENOR_PRECO VARCHAR(1), MAIOR_PRECO VARCHAR(1)").jdbc(url=DB2_Conexao_grupo02_url,
                                                                                                                               table='GRUPO2.TB_SUG_PRECO',
                                                                                                                               mode='overwrite',
                                                                                                                               properties=DATABASE_PROPERTIES)
```
